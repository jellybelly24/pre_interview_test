```python
import numpy as np
from os.path  import join
import struct
#Function to load dataset images and label
def load_image_and_labels(images_path, labels_path):        
        images = []
        with open(images_path, 'rb') as f:
            magic, size, rows, cols = struct.unpack(">IIII", f.read(16))
            images = np.fromfile(f, dtype=np.dtype(np.uint8).newbyteorder('>'))
            images = images.reshape((size, rows, cols))
        
        labels = []
        with open(labels_path, 'rb') as f:
            magic, size = struct.unpack(">II", f.read(8))
            labels = np.fromfile(f, dtype=np.dtype(np.uint8).newbyteorder('>'))
        
        return images, labels
```

```python
#Load dataset
X_train, y_train = load_image_and_labels('./Dataset/train-images.idx3-ubyte', './Dataset/train-labels.idx1-ubyte')
X_test, y_test = load_image_and_labels('./Dataset/t10k-images.idx3-ubyte', './Dataset/t10k-labels.idx1-ubyte')
```

# Model Training

```python
#Perform normalization on training data
X_train_model = np.zeros((28, 28, X_train.shape[0]))
for i in range(X_train.shape[0]):
    X_train_model[:,:,i] = X_train[i,:,:]
x_train_norm = X_train/255
```

## Different amount of convolutional layers

```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models, callbacks

number_of_conv_layers = [1,2,3]
exp_models = []
history=[]

#Creating models with the different number of convolutional layers
for i in range(len(number_of_conv_layers)):    
    model = models.Sequential()
    # model.add(layers.RandomFlip("horizontal_and_vertical"))
    # model.add(layers.RandomRotation(0.4))
    model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
    model.add(layers.BatchNormalization())
    
    for j in range(number_of_conv_layers[i] - 1):
        model.add(layers.MaxPooling2D((2, 2)))
        model.add(layers.Conv2D(2**(5+j+1), (3, 3), activation='relu'))
        model.add(layers.BatchNormalization())
#     model.add(layers.Conv2D(128, (3, 3), activation='relu'))
#     model.add(layers.BatchNormalization())
    model.add(layers.Flatten())
    model.add(layers.Dense(256, activation='relu'))
    model.add(layers.BatchNormalization())
    model.add(layers.Dense(64, activation='relu'))
    model.add(layers.BatchNormalization())
    model.add(layers.Dense(10, activation='softmax'))
    model.compile(optimizer='adam',
                  loss="categorical_crossentropy",
                  metrics=['accuracy'])
    print(model.summary())
    exp_models.append(model)
    checkpoint_filepath = 'checkpoint.weights_dif_cnnlayer' + str(i) + '.h5'
    model_checkpoint_callback = callbacks.ModelCheckpoint(
        filepath=checkpoint_filepath,
        save_weights_only=True,
        monitor='val_accuracy',
        mode='max',
        save_best_only=True)

    history.append( model.fit(x_train_norm, tf.keras.utils.to_categorical(y_train,10), callbacks = [model_checkpoint_callback],  batch_size = 256,epochs=50,validation_split = 0.2))
```

```python
#Plotting the validation accuracy for each configuration when training
for i in range(len(history)):
    plt.plot(history[i].history["val_accuracy"])
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend(["1 Conv layer", "2 Conv layer", "3 Conv layer"])
```

```python
for i in range(len(history)):
    print(str(i+1)+" convolution layer: " + str(np.max(history[i].history["val_accuracy"])))
```

## Different number of dense layers

```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models, callbacks

number_of_dense_layers = [1,2,3,4]
exp_models = []
history=[]

#Creating models with the different number of dense layers
for i in range(len(number_of_dense_layers)):    
    model = models.Sequential()
    # model.add(layers.RandomFlip("horizontal_and_vertical"))
    # model.add(layers.RandomRotation(0.4))
    model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
    model.add(layers.BatchNormalization())
    
    for j in range(2):
        model.add(layers.MaxPooling2D((2, 2)))
        model.add(layers.Conv2D(2**(5+j+1), (3, 3), activation='relu'))
        model.add(layers.BatchNormalization())

    model.add(layers.Flatten())
    for k in range(number_of_dense_layers[i]):
        model.add(layers.Dense(4**(2 + number_of_dense_layers[i] - k), activation='relu'))
        model.add(layers.BatchNormalization())
    
    model.add(layers.Dense(10, activation='softmax'))
    model.compile(optimizer='adam',
                  loss="categorical_crossentropy",
                  metrics=['accuracy'])
    print(model.summary())
    exp_models.append(model)
    checkpoint_filepath = 'checkpoint.weights_dif_denselayer' + str(i) + '.h5'
    model_checkpoint_callback = callbacks.ModelCheckpoint(
        filepath=checkpoint_filepath,
        save_weights_only=True,
        monitor='val_accuracy',
        mode='max',
        save_best_only=True)

    history.append( model.fit(x_train_norm, tf.keras.utils.to_categorical(y_train,10), callbacks = [model_checkpoint_callback],  batch_size = 256,epochs=50,validation_split = 0.2))
```

```python
#Plotting the validation accuracy for each configuration when training
for i in range(len(history)):
    plt.plot(history[i].history["val_accuracy"])
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend(["1 dense layer", "2 dense layer", "3 dense layer", "4 dense layer"])
```

```python
for i in range(len(history)):
    print(str(i+1)+" dense layers: " + str(np.max(history[i].history["val_accuracy"])))
```

## Different batch sizes

```python
batch_sizes = [32,64,128,256,512,1024]
exp_models = []
history=[]

#Creating models with batch sizes
for i in range(len(batch_sizes)):    
    model = models.Sequential()
    # model.add(layers.RandomFlip("horizontal_and_vertical"))
    # model.add(layers.RandomRotation(0.4))
    model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
    model.add(layers.BatchNormalization())
    
    for j in range(2):
        model.add(layers.MaxPooling2D((2, 2)))
        model.add(layers.Conv2D(2**(5+j+1), (3, 3), activation='relu'))
        model.add(layers.BatchNormalization())
#     model.add(layers.Conv2D(128, (3, 3), activation='relu'))
#     model.add(layers.BatchNormalization())
    model.add(layers.Flatten())
    for k in range(2):
        model.add(layers.Dense(4**(4 - k), activation='relu'))
        model.add(layers.BatchNormalization())
    
    model.add(layers.Dense(10, activation='softmax'))
    model.compile(optimizer='adam',
                  loss="categorical_crossentropy",
                  metrics=['accuracy'])
    print(model.summary())
    exp_models.append(model)
    checkpoint_filepath = 'checkpoint.weights_dif_batch_size' + str(batch_sizes[i]) + '.h5'
    model_checkpoint_callback = callbacks.ModelCheckpoint(
        filepath=checkpoint_filepath,
        save_weights_only=True,
        monitor='val_accuracy',
        mode='max',
        save_best_only=True)

    history.append(model.fit(x_train_norm, tf.keras.utils.to_categorical(y_train,10), callbacks = [model_checkpoint_callback],  batch_size = batch_sizes[i],epochs=50,validation_split = 0.2))
```

```python
#Plotting the validation accuracy for each configuration when training

for i in range(len(history)):
    plt.plot(history[i].history["val_accuracy"])
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend(["Batch Size of 32", "Batch Size of 64", "Batch Size of 128", "Batch Size of 256","Batch Size of 512", "Batch Size of 1024"])
```

```python
for i in range(len(history)):
    print(" Batch Size of "+ str(batch_sizes[i]) + ": " + str(np.max(history[i].history["val_accuracy"])))
```

# Model Testing

```python
#Chose the model that trained on a batch size of 128 as the model to perform predictions on the test set.
best_model = exp_models[2]
best_model.load_weights("checkpoint.weights_dif_batch_size128.h5")
y_test_pred = best_model.predict(X_test/255)
```

```python
#Reversing the one hot encoding
y_test_pred = np.array(tf.argmax(y_test_pred, axis = 1))
```

```python
#Obtaining the test accuracy
num_correct = 0
for i in range(y_test_pred.shape[0]):
    if y_test_pred[i] == y_test[i]:
        num_correct += 1
print(num_correct/y_test_pred.shape[0])
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

#Creating a confusion matrix based on predicted classes and the actual classes  
cf_matrix = confusion_matrix(y_test, y_test_pred)
sns.heatmap(cf_matrix, annot=True, annot_kws = {'size': 7})
plt.xlabel('Predicted')
plt.ylabel('Actual')
# sns.set(font_scale  = 0.75)
```
